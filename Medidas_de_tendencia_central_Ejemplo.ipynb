```python
# importar librerias --> pandas
import pandas as pd


# convertir el csv es un dataframe (estructura de datos similar a una tabla con nombres de fila y columnas)
df = pd.read_csv('cars_depurado.csv')
df
```

##Análisis

###Promedio de precios

```python
# PROMEDIO DE PRECIOS --> metodo mean()
promedio= df['price_usd'].mean()
print(promedio)
```

###Mediana de precios

```python
# MEDIANA DE PRECIOS --> metodo median()
mediana= df['price_usd'].median()
print(mediana)
```

**Se puede apreciar una diferencia importante entre el promedio y la mediana de precios --> hay unos pocos autos que tienen precios muy elevados.**





###Frecuencia de precios

```python
# MODA /FRECUENCIA DE PRECIOS
df['price_usd'].plot.hist(bins=20) # graficamos un histograma de 20 columnas
# bins = cantidad de columnas
```

Podemos deducir de este histograma, que la mayoria de los autos tiene un precio menor a USD 10000. (en la proxima unidad aprenderemos a explorar este punto cvon mas detalle y precision)
